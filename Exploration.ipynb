```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
price_history_df = pd.read_csv('input_data.csv')
price_history_df['time'] = pd.to_datetime(price_history_df['time'])
price_history_df = price_history_df.set_index('time')

price_history_df.dtypes
```

```python
price_history_df.describe()
```

```python
price_history_df.loc[price_history_df['prices'] == 9999.0]
```

```python
sns.set()
price_history_df[(price_history_df.index >= '2018-10-01') &
                    (price_history_df.index <= '2018-11-01')]['prices'].plot()
```

```python
price_history_df.loc[price_history_df['prices'] == 9999, 'prices'] = np.nan
```

```python
price_history_df[(price_history_df.index >= '2018-10-01') &
                    (price_history_df.index <= '2018-11-01')]['prices'].plot()
```

```python
price_history_df.plot(title = 'Price History')
```

```python
price_history_df['year'] = pd.DatetimeIndex(price_history_df.index).year
price_history_df['month'] = pd.DatetimeIndex(price_history_df.index).month
price_history_df['week'] = pd.DatetimeIndex(price_history_df.index).week
price_history_df['day'] = pd.DatetimeIndex(price_history_df.index).day
price_history_df['hour'] = pd.DatetimeIndex(price_history_df.index).hour
price_history_df['minute'] = pd.DatetimeIndex(price_history_df.index).minute
price_history_df['minutes_in_hour'] = price_history_df['minute']/60
price_history_df['float_hour'] = price_history_df['hour'] + price_history_df['minutes_in_hour']


price_history_df
```

```python
price_history_df[price_history_df.duplicated()]
```

```python
price_hours = price_history_df.groupby('float_hour').agg('mean')

print(price_hours['prices'].describe())

price_hours['prices'].plot()
```

```python
price_hours = price_history_df.groupby('hour').agg('mean')

fig, ax = plt.subplots(figsize = (8, 4))
ax.set_title('Mean Electricity Prices by Hour (2016-2023)')
ax.set_ylabel('Price (Â£/MWh)')

print(price_hours['prices'].idxmax())
print(price_hours['prices'].idxmin())

price_hours['prices'].plot(ax = ax)

ax.set_xlabel('Hour of Day')
```

```python
price_hours = price_history_df.groupby(['year', 'float_hour']).agg('mean').reset_index().set_index('float_hour')

price_hours_by_yr = price_hours.groupby('year')

fig, ax = plt.subplots(figsize = (8, 4))
ax.set_title('Mean Electricity Prices by Hour by Year')
ax.set_ylabel('Price (Â£/MWh)')


for idx, (year, df) in enumerate(price_hours_by_yr):
    df['prices'].plot(label = year, ax = ax, color = c[idx])

ax.set_xlabel('Hour of Day')
ax.legend()
```

```python
price_hours = price_history_df.groupby(['year', 'float_hour']).agg('mean').reset_index().set_index('float_hour')

price_hours_by_yr = price_hours.groupby('year')

fig, ax = plt.subplots(figsize = (20, 10))
ax.set_title('Mean Hourly Electricity Prices by Year')
ax.set_ylabel('Price (Â£/MWh)')

c = ['r', 'g', 'b', 'black', 'orange', 'purple', 'darkgray', 'darkgreen']


for idx, (year, df) in enumerate(price_hours_by_yr):
    sns.histplot(df['prices'], label = year, ax = ax, binwidth = 5, color = c[idx], alpha = 0.3)

ax.set_xlabel('Hour of Day')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize = (8, 4))
ax.set_title('Distribution of Electricity Prices by Year')

sns.histplot(price_hours, x = 'prices', hue = 'year', palette = c, binwidth = 5, multiple = 'stack', )

ax.set_xlabel('Price (Â£/MWh)')
```

```python
price_hours = price_history_df.groupby(['month', 'float_hour']).agg('mean').reset_index().set_index('float_hour')

price_hours_by_yr = price_hours.groupby('month')

fig, ax = plt.subplots(figsize = (20, 10))
ax.set_title('Mean Hourly Electricity Prices by Month')
ax.set_ylabel('Price (Â£/MWh)')


for year, df in price_hours_by_yr:
    df['prices'].plot(label = year, ax = ax)

ax.set_xlabel('Hour of Day')
ax.legend()
```
